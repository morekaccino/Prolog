```python
words = '''markus
S
s
Markus
1990
york
York
lea
Lea
tiana
Tiana
2018
2014
2015'''
words = words.split()
words
```

```python
from itertools import permutations, combinations
perm = combinations(words, 1)
list(perm)
for i in range(1, len(words) + 1):
  perm = list(combinations(words, i))
  for j in perm:
    s = ""
    for k in j:
      s += k
    
    if len(s) < 12:
      print(s)
```
