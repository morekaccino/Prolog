```python
words = '''markus
1990
york
lea
tiana
2018
2014
2015'''
words = words.split()
words
```

```python
from itertools import permutations, combinations
perm = combinations(words, 1)
list(perm)
for i in range(1, len(words) + 1):
  perm = list(combinations(words, i))
  for j in perm:
    for k in j:
      print(k, end="")
    print()
    
    
  
  
```
